### This following function test whether the partition i and j can be merged. If it can be merged, then we add j to be i's children and also update i's neighbours' list by concatenating j's neighbours' list 

```python
def test_merge(i,j):
    
    #print "test_merge_start",i,j
    numEigen_i = bigdata_list[i][2]
    numEigen_j = bigdata_list[i][2]
    
    numCount_i = bigdata_list[i][3]
    numCount_j = bigdata_list[j][3]
    total_count = numCount_i + numCount_j
    
    mean_i = bigdata_list[i][8:8+365]
    mean_j = bigdata_list[j][8:8+365]
    
    cov_i = bigdata_list[i][8+365:]
    cov_j = bigdata_list[j][8+365:]
    
    #print size(cov_i)
    cov_i1 = np.array(cov_i)
    cov_j1 = np.array(cov_j)
    cov_i1 = cov_i1.reshape(365,365)
    cov_j1 = cov_j1.reshape(365,365)
    
    matrix_i = cov_i1*numCount_i
    matrix_j = cov_j1*numCount_j
    
    matrix = (matrix_i+matrix_j)/ (numCount_i+numCount_j)
    
    mean = np.array(mean_i)*numCount_i + np.array(mean_j)*numCount_j
    mean = mean/ (numCount_i+numCount_j)

    U,D,V=np.linalg.svd(matrix)
    total = sum(D)
    count = 0
    n=0
    for n in range(0,730):
        count+=D[n]
        if count/total>0.95:
            break
    n = n+1
    print n,numEigen_i,numEigen_j
    
    if n < numEigen_i + numEigen_j+30:
        #cov,mean,count,n
        partitionId_i = bigdata_list[i][1]
        partitionId_j = bigdata_list[j][1]
        
        matrix = matrix.reshape(1,365*365)
        matrix = matrix.tolist()
        bigdata_list[i] = [bigdata_list[i][0],bigdata_list[i][1],n,total_count,0,0,0,0] + mean.tolist() + matrix[0]
        print 'test_merge_succ',i,n,numEigen_i,numEigen_j
        
        
        neighbour[partitionId_i] = neighbour[partitionId_i] + neighbour[partitionId_j]
        parent[partitionId_j] = partitionId_i

        return 1
    else:
        return 0
```

### This function check whether the two partitions are neighbours.

```python
def check_is_neighbour(i,j):
    id_i = bigdata_list[i][1]
    id_j = bigdata_list[j][1]
    parent_i = id_i
    parent_j = id_j
    if parent_i >= 1194 or parent_j >= 1194:
        return False
    while parent[parent_i]!=-1:
        parent_i = parent[parent_i]
    while parent[parent_j]!=-1:
        parent_j = parent[parent_j]
    
    if parent_i == parent_j:
        return False
    return parent_j in neighbour[parent_i]
```

### This is the main function to run merge code. This code is running for many time. Each time can be seen as an iteration.

```python
length = 1194
for i in range(0,length):
    for j in range(i+1,length):
        if check_is_neighbour(i,j)==True:
            result = test_merge(i,j)
            if result == 1:
                print 'success',i,j,result
```

### This code assign all the original partitions a new partitionId, which is the partitionId of the root of the tree.

```python
numPoint = 1194
IdAfterPartition = np.zeros(numPoint).tolist()
for i in range(0,numPoint):
    index = parent[i]
    if index == -1:
        IdAfterPartition[i] = i
    else:
        while parent[index]!= -1:
            index = parent[index]
        IdAfterPartition[i] = index
```

### Then the file 'new_partition_Id' records the original and latest partition id for each partition.

```python
out = open('new_partition_Id','w')

for i in range(0,1194):
    string = str(IdAfterPartition[i]) + ' ' + str(bigdata[0][i]) +'\n'
    print string
    
    out.write(string)

out.close()
```
