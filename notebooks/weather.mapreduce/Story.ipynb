 **<h3>First Step: Data Preprocessing</h3>**

In order to have compact and clean dataframe that contains all the features that might be needed in the future, we preprocess the ncdc data and concatenate selected features of measurements for each (station,year) pair. We select 13 features that *we* think that might be helpful in the future. We deliberately deleted the segments of data source counts and other hard-to-interpret features.
We choose the following 13 attributes available in the ncdc dataset and below we summerize the brief description of the features and our plan of how to use this data.

1. TEMP: Mean temperature for the day in degrees Fahrenheit to tenths.Missing = 9999.9 (Celsius to tenths for metric version.)
2. DEWP: Mean dew point for the day in degrees Fahrenheit to tenths. Missing = 9999.9 (Celsius to tenths for metric version.)
3. SLP: Mean sea level pressure for the day in millibars to tenths. Missing = 9999.9
4. STP: Mean station pressure for the day in millibars to tenths. Missing = 9999.9
5. VISIB: Mean visibility for the day in miles to tenths. Missing = 999.9 (Kilometers to tenths for metric version.)
6. WDSP: Mean wind speed for the day in knots to tenths. Missing = 999.9 (Meters/second to tenths for metric version.)
7. MXSPD: Maximum sustained wind speed reported for the day in knots to tenths.Missing = 999.9
8. GUST: Maximum wind gust reported for the day in knots to tenths. Missing = 999.9 (Meters/second to tenths for metric version.)
9. MAX: Maximum temperature reported during the day in Fahrenheit to tenths--time of max temp report varies by country and region, so this will sometimes not be the max for the calendar day. Missing = 9999.9
10. MIN: Minimum temperature reported during the day in Fahrenheit to tenths--time of min temp report varies by country and region, so this will sometimes not be the min for the calendar day. Missing = 9999.9
11. PRCP: Total precipitation (rain and/or melted snow) reported during the day in inches and hundredths; will usually not end with the midnight observation--i.e., may include latter part of previous day. Missing = 99.99
12. SNDP：Snow depth in inches to tenths--last report for the day if reported more than once. Missing = 999.9 (Centimeters to tenths for metric version.) Note: Most stations do not report '0' on days with no snow on the ground--therefore,'999.9' will often appear on these days.
13. FRSHTT: Indicators (1 = yes, 0 = no/not reported) for the occurrence during the day of:
     Fog ('F' - 1st digit).
     Rain or Drizzle ('R' - 2nd digit).
     Snow or Ice Pellets ('S' - 3rd digit).
     Hail ('H' - 4th digit).
     Thunder ('T' - 5th digit).
     Tornado or Funnel Cloud ('T' - 6th digit).

**<h3>Model Merging</h3>**

In the previous section, we have computed the mean vector and covariance matrix for each partition and have performed the complete PCA for a single partition. Our final task now is to run the PCA for all partitions and merge them, using the medium description length (MDL) as a measure. 
The MDL criterion for when to merge two regions 1 and 2, to a new region 3 is
n1⋅k1+(k1+1)⋅(2×365)+n2⋅k2+(k2+1)⋅(2×365)>n3⋅k3+(k3+1)⋅(2×365)

where ki is the number of required eigenvectors for region i to explain 99% of the variance and ni are the number of measurements in region i. In order to do that, we will proceed as follows: Find neighbors for each partition; Store neighbor information in a graph; Merge partitions based on MDL
