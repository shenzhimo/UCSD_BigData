### **1, partition the world into small parts**  For our partition, we first divide the world using latitude/longitude for many iterations.  Then for each partition, we then partition by medium value of longitude/latitude.

```python
import pandas as pd
colspecs = [(1, 7), (10, 17), (20, 28), (31, 38)]
Data = pd.read_fwf('join_data_1',colspecs =colspecs,  header=None,
                       names=['station','latitude','longitude','count'])

Data['latitude1']=-90000
Data['latitude2']=91000
Data['longitude1']=-180000
Data['longitude2']=181000
Data['area']=0
Data.head()

#print len(Data)
for i in range(0,len(Data)):
    Data['station'][i]=Data['station'][i].replace('"','')
    Data['latitude'][i]=Data['latitude'][i].replace('"','')
    Data['latitude'][i]=int(Data['latitude'][i])
    Data['longitude'][i]=Data['longitude'][i].replace('"','')
    Data['longitude'][i]=int(Data['longitude'][i])
    Data['count'][i]=Data['count'][i].replace('"','')
    Data['count'][i]=Data['count'][i].replace(']','')
    Data['count'][i]=int(Data['count'][i])

djoin=Data
djoin.head()

def mypartition(start,end,index,l,d):
    if end-start<=1:
        return l
    if index=='latitude':
        index2='longitude'
    else:
        index2='latitude'
    a=djoin[start:end].sort(index,inplace=False)
    djoin[start:end]=a
    temp=sum(djoin[start:end]['count'])
    #print temp
    if temp<300 :
        return l
    num=0
    if d<8:
        t1=(djoin.ix[start,[index+'1']].values+djoin.ix[start,[index+'2']].values)/2
        for i in range(start,end):
            if djoin.ix[i,[index]].values>t1:
                djoin.ix[start:i,[index+'2']]=t1
                #l=l+1
                l=mypartition(start,i,index2,l,d+1)
                djoin.ix[i:end,['area']]=l
                djoin.ix[i:end,[index+'1']]=t1
                l=l+1
                l=mypartition(i,end,index2,l,d+1)
                return l
        return l
    for i in range(start,end):
        #print num
        #print djoin.ix[i,['count']]
        num=num+djoin.ix[i,['count']].values
        if num>temp/2:
            if start==i:
                #djoin.ix[start:i+1,['area']]=l
                t1=(djoin.ix[i,[index]].values+djoin.ix[i+1,[index]].values)/2
                #print t1
                djoin.ix[start:i+1,[index+'2']]=t1
                #l=l+1
                l=mypartition(start,i+1,index2,l,d+1)
                djoin.ix[i+1:end,['area']]=l
                djoin.ix[i+1:end,[index+'1']]=t1
                l=l+1
                l=mypartition(i+1,end,index2,l,d+1)
                return l
            else:
                #djoin.ix[start:i,['area']]=l
                t1=(djoin.ix[i,[index]].values+djoin.ix[i-1,[index]].values)/2
                #print t1
                djoin.ix[start:i,[index+'2']]=t1
                #l=l+1
                l=mypartition(start,i,index2,l,d+1)
                djoin.ix[i:end,['area']]=l
                djoin.ix[i:end,[index+'1']]=t1
                l=l+1
                l=mypartition(i,end,index2,l,d+1)
                return l
```
